**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import pandas as pd
import numpy as np
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
print(df)
````

# Diabetes_Model


````python
import pandas as pd
df=pd.read_csv("/content/drive/MyDrive/ICT/diabetes_model.csv")
````

````python
df.head()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
plt.figure(figsize=(8, 6))
plt.hist(df['Pregnancies'], bins=5, color='red')
plt.title('Histogram of Pregnancies')
plt.xlabel('Pregnancies')
plt.ylabel('Frequency')
plt.show()
````

The histogram for the 'Pregnancies' column will provide a visual representation of the distribution of Pregnancies in the dataset. By setting 5 bins, the histogram will group the pregnancies data into 5 intervals, allowing you to see the frequency of pregnancie case among individuals in each range. The color red gives a distinct look. This plot can help identify patterns, such as pregnancies among different age group and the ratio of pregnancies.

````python
plt.figure(figsize=(8, 6))
plt.hist(df['Glucose'], bins=5, color='purple')
plt.title('Histogram of Glucose')
plt.xlabel('Glucose')
plt.ylabel('Frequency')
plt.show()
````

The histogram for the 'Glucose' column will provide a visual representation of the distribution of glucose levels among the individuals in the dataset. By setting 5 bins, the histogram will group the glucose data into 5 intervals, allowing you to see the frequency of glucose levels in each range. The color "violet" gives a distinct look, while the black edges make the bins easy to distinguish. This plot can help identify patterns, such as whether glucose levels are skewed towards higher or lower values, or if they follow a normal distribution, which could be critical in understanding diabetes risk factors in this population.

````python
# df.plot.scatter(x='BMI', y='Age', title='BMI vs. Age')
sns.scatterplot(x = 'Age', y = 'BMI', hue = 'Outcome', data = df)
plt.title('BMI vs. Age')
plt.show()
````

 The x-axis represents the “Age” and the y-axis represents the “BMI” (Body Mass Index).The hue parameter is specified as 'Outcome'. This means that the scatter plot will use the “Outcome” column from the DataFrame (df) to color-code the data points.The colors on the scatter plot will correspond to different outcomes. For example, if “Outcome” has two values (e.g., “Age” and “BMI”), the plot will use distinct colors for each group.


````python
sns.scatterplot(x = 'Insulin', y = 'Glucose', hue = 'Outcome', data = df)
plt.title('Insulin vs. Glucose')
plt.xlabel('Insulin')
plt.ylabel('Glucose')
plt.legend()
plt.show()
````

````python
sns.pairplot(df[['BMI', 'Age','BloodPressure']])
````

Diagonals: Each diagonal plot shows the distribution of individual variables. For example, "BloodPressure" shows a fairly wide spread, while "Age" appears to have a broader range with some concentration in middle age.

Scatterplots: The off-diagonal scatterplots highlight the pairwise relationships. There seems to be no strong linear relationship between these variables at first glance. For instance, "BloodPressure" and "Age" exhibit a slight positive correlation.
