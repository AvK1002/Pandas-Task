```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
plt.figure()
sns.violinplot(x='species', y='sepal_length', data=iris)
plt.title('Distribution of Sepal Length by Species')
plt.show()
```

```python
plt.figure()
sns.swarmplot(x='species', y='sepal_width', data=iris)
plt.title('Sepal Width Distribution by Species')
plt.show()
```

```python
plt.figure()
sns.stripplot(x='species', y='petal_length', data=iris)
plt.title('Petal Length Distribution by Species')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='species', y='petal_width', data=iris)
plt.title('Petal Width Distribution by Species')
plt.show()
```

```python
plt.figure()
sns.barplot(x='species', y='sepal_length', data=iris)
plt.title('Mean Sepal Length by Species')
plt.show()
```

```python
plt.figure()
sns.countplot(x='species', data=iris)
plt.title('Count of Each Species')
plt.show()
```

```python
plt.figure()
sns.heatmap(iris.drop('species', axis=1).corr(), annot=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure()
sns.barplot(x='species', y='petal_width', hue='sepal_length', data=iris)
plt.title('Mean Petal Width by Species and Sepal Length')
plt.show()
```

```python
plt.figure()
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs. Sepal Width')
plt.show()
```

```python
plt.figure()
sns.lineplot(x='species', y='petal_length', data=iris)
plt.title('Trend of Petal Length Over Species')
plt.show()
```
